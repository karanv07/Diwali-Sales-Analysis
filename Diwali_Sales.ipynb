```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Diwali Sales Data.csv", encoding = "unicode_escape")
df.head()
```

```python
df.info()
```

```python
df.drop(["Status", "unnamed1"], axis = 1, inplace = True)
```

```python
df.info()
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
df["Amount"] = df["Amount"].astype("int")
```

```python
df["Amount"].dtypes
```

```python
def conv(value):
    if value == 0:
        return "Married"
    else:
        return "Unmarried"
df["Marital_Status"] = df["Marital_Status" ].apply(conv)
```

```python
df.columns
```

```python
df[["Age", "Orders", "Amount"]].describe()
```

# Exploratory Data Analysis

## Gender

```python
plt.figure(figsize = (4,4.5))
ax = sns.countplot(x = "Gender", data = df, hue = "Gender", palette = "Set2", legend = False)
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title("Count of Genders")
plt.show()
```

```python
plt.figure(figsize = (4,4))
gb = df.groupby(["Gender"], as_index = False)["Amount"].sum().sort_values("Amount", ascending = False)
ax = sns.barplot(x = "Gender", y = "Amount", data = gb, hue = "Gender", palette = "Set2", legend = False)
ax.bar_label(ax.containers[0], fmt = "%.0f")
ax.bar_label(ax.containers[1], fmt = "%.0f")
plt.title("Amount by Gender")
plt.show()
```

```python
## from above charts we can see that most of the buyers are Females and purchasing power of female is more than men
```

## Age Group

```python
plt.figure(figsize = (8,4))
ax = sns.countplot(x = "Age Group", data = df, hue = "Gender", palette = "Set2")
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title("Gender wise Age Group")
plt.show()
```

```python
plt.figure(figsize = (7,4))
gb = df.groupby(["Age Group"], as_index = False)["Amount"].sum().sort_values("Amount", ascending = False)
sns.barplot(x = "Age Group", y = "Amount", data = gb, hue = "Age Group", palette = "Set2", legend = False)
plt.title("Amount by Age Group")
plt.show()
```

```python
## From the above graphs we can see that age group between 26-35 are likely more buyers especially females.
```

## State

```python
plt.figure(figsize = (10,5))
gb = df.groupby(["State"], as_index = False)["Orders"].sum().sort_values("Orders", ascending = False).head(10)
ax = sns.barplot(x = "State", y = "Orders", data = gb, hue = "State", palette = "Set2", legend = False)
for container in ax.containers:
    ax.bar_label(container)
plt.title("Top 10 States by Orders")
plt.xticks(rotation = 30)
plt.show()
```

```python
plt.figure(figsize = (10,5))
gb = df.groupby(["State"], as_index = False)["Amount"].sum().sort_values("Amount", ascending = False).head(10)
ax = sns.barplot(x = "State", y = "Amount", data = gb, hue = "State", palette = "Set2", legend = False)
for container in ax.containers:
    ax.bar_label(container, fmt = "%.0f")
plt.xticks(rotation = 30)
plt.title("Top 10 States by Amount")
plt.show()
```

```python
## Uttar Pradesh, Maharasthra and Karnataka contributes more orders and amount compared to other states.
```

## Marital Status

```python
plt.figure(figsize = (3,4))
ax = sns.countplot(x = "Marital_Status", data = df, hue = "Marital_Status", palette = "Set2", legend = False)
for container in ax.containers:
    ax.bar_label(container)
plt.title("Marital Status")
plt.show()
```

```python
gb = df.groupby(["Marital_Status", "Gender"], as_index = False)["Amount"].sum().sort_values("Amount", ascending = False)
plt.figure(figsize = (3,3))
ax = sns.barplot(x = "Marital_Status", y = "Amount", data = gb, hue = "Gender", palette = "Set2")
plt.show()
```

## Occupation

```python
plt.figure(figsize = (17,4))
ax = sns.countplot(x = "Occupation", data = df, hue = "Occupation", palette = "Set2", legend = False)
for bars in ax.containers:
    ax.bar_label(bars)
plt.xticks(rotation = 30)
plt.show()
```

```python
gb = df.groupby(["Occupation"], as_index = False)["Amount"].sum().sort_values("Amount", ascending = False)
plt.figure(figsize = (17,4))
sns.barplot(x = "Occupation", y = "Amount", data = gb, hue = "Occupation", palette = "Set2", legend = False)
plt.xticks(rotation = 30)
plt.show()
```

```python
# IT sector, Healthcare and Aviation are the top 3 occupations.
```

## Product Category

```python
plt.figure(figsize = (18,4))
ax = sns.countplot(x = "Product_Category", data = df, hue = "Product_Category", palette = "Set2")
for bars in ax.containers:
    ax.bar_label(bars)
plt.xticks(rotation = 50)
plt.show()
```

```python
plt.figure(figsize = (18,4))
gb = df.groupby(["Product_Category"], as_index = False)["Amount"].sum().sort_values("Amount", ascending = False)
sns.barplot(x = "Product_Category", y = "Amount", data = gb, hue = "Product_Category", palette = "Set2", legend = False)
plt.xticks(rotation = 50)
plt.show()
```

```python
# Food, Clothing & Apparel and Electronics & Gadgets are top 3 Product categories which contributs more amount compared to other categories.
```
